## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
# your code here
nsq=10
sqn = [sqnumber**2 for sqnumber in range(nsq+1)]
print(sqn)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
# your code here

nsq=10
sqn = [sqnumber**2 for sqnumber in range(nsq+1) if sqnumber%2==1]

print(sqn)
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
# your code here

nsq=1000
sqn = [sqnumber**2 for sqnumber in range(nsq) if sqnumber%8==0]

print(sqn)
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
# your code here
len(people)
```

**How many people have kids**?

```python
# your code here
count=0
for parent in people:
    if int(parent["n_kids"])>0:
        count+=1

print(count)

    
```

**How many kids do they have in total?**

```python
# your code here
countkids=0

for parent in people:
    countkids+=parent["n_kids"]


print(countkids)
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
# your code here

peopleyear=people


for peopley in peopleyear:
    namelast=peopley["name"]
    if namelast[-1]=="a":
        peopley["n_kids"]+=1

print(peopleyear)
```
